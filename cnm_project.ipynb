```
import string
description={}
def splitting(x):
  table=str.maketrans('','',string.punctuation)
  i=[]
  i=x.split()
  i=[word.lower() for word in i]
  i=[w.translate(table) for w in i]
  i=[w for w in i if w!='']
  return i
```

```
#This will deal with query
import xml.etree.ElementTree as ET
file_query = open('en.topics.76-125.2010.txt','r')
tree=ET.parse(file_query)
root=tree.getroot()
query_no=[]
query_desc=[]
query_desc_2=[]
for top in root.findall('top'):
  query_no.append(top.find('num').text)
  query_desc.append(top.find('desc').text)
  query_desc_2.append(top.find('desc').text)

i=0
for doc in query_desc:
  query_desc[i]=splitting(doc)
  i+=1    
```

```
#this is for query expansion for tfidf
import nltk
nltk.download('wordnet')
from nltk.corpus import wordnet
k=0
for query in query_desc:
  for word in query:
    temp_syn=[]
    for syn in wordnet.synsets(word):
      temp=syn.name().split('.')
      if(temp[1] == 'n'):
        if(temp[0] not in temp_syn and temp[0]!= word):
          temp_syn.append(temp[0])
        if(len(temp_syn) == 3):
          break
  for extra in temp_syn:
    query.append(extra)
  print("  ")
  query_desc_2[k]=query
  k+=1 
```

```

# Python program to convert a list to string 
    
# Function to convert   
def listToString(s):  
    
    # initialize an empty string 
    str1 = ""  
    
    # traverse in the string
    k=0   
    for ele in s:
      if(k!=0):
        str1+=" "
        str1+=ele  
      else:
         str1+=ele
      k+=1
    # return string   
    return str1  
        
        
# Driver code   
k=0
query_desc_3=[]
for query in query_desc_2:
  query_desc_3.append(listToString(query))
```

```
#It will read the actual relevant query
relevance_file=open('en.qrels.76-125.2010.txt','r')
relevant_temp=[]
line_temp=[]
for line in relevance_file:
  line_temp=line.split(' ')
  line_temp[3]=line_temp[3].split()   #I am doing this in order to remove \n from '0\n'
  relevant_temp.append(line_temp)
```

```
#This will calculate the actual relevant document
import numpy as np
from collections import defaultdict
actual_relevant = defaultdict(str)   #This will give me the name of the document which are relevant to the given query
for line in relevant_temp:
  if(line[3][0]=='1'):
    actual_relevant[line[0]] +=line[2]
    actual_relevant[line[0]] +=' '
```

```
#This will print the relevant document calculated and actual document document that is relevant to the query with query expansion
k=0
irrelevant=[]
intersect_tfidf=[]  #This will get the intersection b/w the document calculated by us wrt to tha actual document
for key in actual_relevant.keys():   #Because the number of query are from range 76 to 125
  print('For query no.',key)
  our=top_relevant_tfidf[k]
  
  k+=1
  print('Our top five=',our)
 # for doc in our:      #I am doing this in order to get irrelevant document

  actual=actual_relevant[key].split(' ')
  print('Actual relevant document',actual)
  intersect_temp=[]
  for doc1 in our:
    for doc2 in actual:
      if(doc1 == doc2):
        intersect_temp.append(doc1)
  irrelevant.append(len(our)-len(intersect_temp)) #I am doing this in order to get irrelevant document
  intersect_tfidf.append(intersect_temp)
```

```
#This will calculate precision score,recall score and f1 score
k=0
p_score=0  #For precision
r_score=0  #for recall score
f_score=0  #for f1 score
for i in range(76,126):
  if(len(intersect_tfidf[k])!=0):
    tp=len(intersect_tfidf[k])  #True positive
    print(tp)
    fp=15-tp
    actual=actual_relevant[i].split(' ')
    fn=len(actual)-tp
    precision=tp/(fp+tp)
    recall=tp/(fn+tp)
    f1=2*((precision*recall)/(precision+recall))
    print('For query no=',i,'precision score=',precision,'recall score=',recall,'F1 score=',f1)
    p_score +=precision
    r_score +=recall
    f_score +=f1

  k=k+1
```

```
print('Score before query expansion','precision score=',p_score/k,'recall score=',r_score/k,'F1 score=',f_score/k)
```

```
#This to get optimal query for tfidf+query expansion+rochio+gamm=0.15
query_optimal_tfidf=[]
length=len(corpus)
D_ir=length-15
k=0
for query in query_tfidf:
  relevant_q=svd.transform(query_relevant[k])
  query_irrelevant=sum_all-relevant_q
  query_opt=query+(0.85*(relevant_q/15))-(0.15*(query_irrelevant/D_ir))
  query_optimal_tfidf.append(query_opt)
  k+=1
```
